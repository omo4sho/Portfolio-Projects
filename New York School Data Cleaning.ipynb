```python
import pandas as pd
data_files = [
    "ap_2010.csv",
    "class_size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]
data = {}
```

```python
data_directory = ""

# Loop through the data files and read them into dataframes
for file_name in data_files:
    # Construct the full file path
    file_path = data_directory + file_name
    
    # Extract the base filename without the extension (e.g., "ap_2010")
    base_filename = file_name.split(".")[0]
    
    # Read the CSV file into a dataframe and store it in the dictionary
    data[base_filename] = pd.read_csv(file_path)
```

```python
#Inserting DBN Fields
# Copy the dbn column in hs_directory into a new column called DBN
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

# Define a custom function to pad the CSD values
def pad_csd(csd):
    csd_str = str(csd)
    if len(csd_str) == 1:
        return "0" + csd_str
    else:
        return csd_str

# Apply the custom function to generate the padded_csd column in class_size
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(pad_csd)

# Combine padded_csd and SCHOOL CODE columns to create the DBN column in class_size
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]

# Display the first few rows of class_size to check the DBN column
print(data["class_size"].head())
```

```python
# Convert SAT score columns to numeric with errors='coerce'
sat_columns = ["SAT Math Avg. Score", "SAT Critical Reading Avg. Score", "SAT Writing Avg. Score"]
for col in sat_columns:
    data["sat_results"][col] = pd.to_numeric(data["sat_results"][col], errors="coerce")

# Create the sat_score column by adding up the SAT scores
data["sat_results"]["sat_score"] = data["sat_results"][sat_columns].sum(axis=1)

# Display the first few rows of the sat_score column to verify
print(data["sat_results"]["sat_score"].head())
```

```python
#Parsing Geographic Coordinates for Schools
import re

def extract_latitude(coordinates_str):
    # Define a regular expression pattern to match latitude and longitude coordinates
    pattern = r"\(([-\d.]+), ([-\d.]+)\)"
    
    # Use re.search() to find the first match in the string
    match = re.search(pattern, coordinates_str)
    
    # Check if a match is found
    if match:
        # Extract the latitude (group 1 in the regex pattern)
        latitude = float(match.group(1))
        return latitude
    else:
        # Return None if no coordinates are found
        return None

# Example usage:
coordinates_str = "Coordinates: (40.8276026690005, -73.90447525699966)"
latitude = extract_latitude(coordinates_str)
if latitude is not None:
    print("Latitude:", latitude)
else:
    print("No latitude found in the string.")
    
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(extract_latitude)

# Display the first few rows of hs_directory to verify the results
print(data["hs_directory"].head())
```

```python
#Extracting the Longitude
# Apply the extract_longitude function to the Location 1 column and assign the result to the lon column
import re
def extract_longitude(coordinates_str):
    # Define a regular expression pattern to match latitude and longitude coordinates
    pattern = r"\(([-\d.]+), ([-\d.]+)\)"
    
    # Use re.search() to find the first match in the string
    match = re.search(pattern, coordinates_str)
    
    # Check if a match is found
    if match:
        # Extract the longitude (group 2 in the regex pattern)
        longitude = float(match.group(2))
        return longitude
    else:
        # Return None if no coordinates are found
        return None

# Apply the extract_longitude function to the Location 1 column and assign the result to the lon column
data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(extract_longitude)

# Convert the lat and lon columns to numbers with errors="coerce"
data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors="coerce")
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors="coerce")

# Display the first few rows of hs_directory to verify the results
print(data["hs_directory"].head())
```
